```python
import numpy as np
import pandas as pd
```

```python
data = {'A':[1,2,np.nan],'B':[15,np.nan,np.nan],'C':[1,2,3]}
```

```python
df = pd.DataFrame(data)
df
```

```python
df.dropna(axis=1)
```

```python
df.dropna(thresh=2)
```

```python
df.fillna(value="Fill Value")
```

```python
df['A'].fillna(value=df['A'].mean())
```
